```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib as plot
import math
import talib
from sklearn import metrics, linear_model, preprocessing, svm
```

```python
conn = sqlite3.connect('poloniex_0.1.db')
c = conn.cursor()
```

```python
df = pd.read_sql_query("SELECT * FROM candles_USDT_BTC", conn)
df = df.drop('id' ,1)
df['start'] = pd.to_datetime(df['start'],unit='s')
df.set_index('start' , inplace = True)
#df.index.dtype
df.head(10)
#print (len(df.index))
```

```python
def ohlcsum(df):
    #df = df.sort()
    return {
       'open': df['open'][0],
       'high': df['high'].max(),
       'low': df['low'].min(),
       'close': df['close'][-1],
       'volume': df['volume'].sum(),
        'trades' : df['trades'].sum(),
        'vwp': (df['vwp']* df['volume']).sum()/df['volume'].sum()
        
      }
```

```python
def convertcandles(df, mins):
    s = str(mins) + 'T'
    return( df.resample(s).agg(ohlcsum))
    
    
#convertcandles(df, 120)
    
```

```python
newdf = convertcandles(df,1500)

#newdf = df
price = np.array(newdf['vwp'])
sma  = talib.SMA(price,50)
m = talib.MACD(price, 12,30,9)
#m[1][40:60]
newdf.head(20)
```
